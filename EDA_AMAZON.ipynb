## IMPORTING THE LIBRARIES

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## TYPECASTING THE STRING VALUES INTO FLOAT VALUES FOR PROPER VISUALIZATION

```python
df['actual_price'] 
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹' ,"")
df['actual_price'] = df['actual_price'].str.replace(',' ,"")
```

```python
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['actual_price']
```

```python
df['discounted_price']
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹' ,"")
df['discounted_price'] = df['discounted_price'].str.replace(',' ,"")
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['discounted_price']
```

```python
df['discount_percentage']
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%","")
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

### FINDING THE CORRELATION WITH THE ACTUAL PRICE vs DISCOUNTED PRICE vs DISCOUNT PERCENT OF THE PRODUCT INTO TABLE 1

```python
tab1 = df[['actual_price','discounted_price','discount_percentage']].corr()
tab1
```

### FORMING THE HEATMAP , DISTPLOT & LINE PLOT 

```python
sns.heatmap(tab1)
```

```python
sns.distplot(tab1)
```

```python
plt.plot(tab1)
plt.legend(["actual_price",'discounted_price','discount_percentage'])
plt.title("Correlation values of Table 1")
plt.show()
```
